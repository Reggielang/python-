```python
#将暂无资料的数据替代为缺失值（Nan）
import pandas as pd 
df = pd.read_excel('C:/Users/REGGIE/Desktop/数据分析资料/data/house_sample.xlsx')
df.head()
```

```python
import numpy as np
def convert(a):
    if a == '暂无资料':
        return np.nan
    else:
        return a
    
df.applymap(convert)
```

```python
df.applymap(lambda a : np.nan if a == '暂无资料' else a)
```
