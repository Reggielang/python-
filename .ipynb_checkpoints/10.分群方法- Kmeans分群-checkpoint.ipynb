```python
from sklearn.datasets import load_iris
iris = load_iris()

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, init = 'k-means++',random_state=123)
y_kmeans = kmeans.fit_predict(iris.data)
```

```python
y_kmeans
```

```python
#0,2, 以及 0，3 0代表标签，数据集的花瓣长度和宽度为X轴和Y轴，所以为2和3
import matplotlib.pyplot as plt

plt.scatter(iris.data[y_kmeans ==0,2],iris.data[y_kmeans == 0,3], s = 100, c ='red',label = 'C1')
plt.scatter(iris.data[y_kmeans ==1,2],iris.data[y_kmeans == 1,3], s = 100, c ='blue',label = 'C2')
plt.scatter(iris.data[y_kmeans ==2,2],iris.data[y_kmeans == 2,3], s = 100, c ='green',label = 'C3')
plt.title('Clusters of iris')
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.legend()
plt.show()
```

```python
#分为三个群，然后三个群的中心
kmeans.cluster_centers_
```

```python
plt.scatter(iris.data[y_kmeans ==0,2],iris.data[y_kmeans == 0,3], s = 100, c ='red',label = 'C1')
plt.scatter(iris.data[y_kmeans ==1,2],iris.data[y_kmeans == 1,3], s = 100, c ='blue',label = 'C2')
plt.scatter(iris.data[y_kmeans ==2,2],iris.data[y_kmeans == 2,3], s = 100, c ='green',label = 'C3')

plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,3],s = 100, c = 'black', label = 'center')
plt.title('Clusters of iris')
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.legend()
plt.show()
```

```python
import pandas as pd

dataset = pd.read_csv('C:/Users/REGGIE/Desktop/数据分析资料/data/customers.csv')
x = dataset.iloc[:,3:5].values
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, init='k-means++',random_state=42)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans ==0,0],x[y_kmeans == 0,1], s = 100, c ='red',label = 'C1')
plt.scatter(x[y_kmeans ==1,0],x[y_kmeans == 1,1], s = 100, c ='magenta',label = 'C2')
plt.scatter(x[y_kmeans ==2,0],x[y_kmeans == 2,1], s = 100, c ='green',label = 'C3')
plt.scatter(x[y_kmeans ==3,0],x[y_kmeans == 3,1], s = 100, c ='yellow',label = 'C4')
plt.scatter(x[y_kmeans ==4,0],x[y_kmeans == 4,1], s = 100, c ='cyan',label = 'C5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s = 100, c = 'black', label = 'center')

plt.title('Clusters of customers')
plt.xlabel('income($)')
plt.ylabel('spending score(1-100)')
plt.legend()
plt.show()
#针对不同的群组进行不同的客户分类，然后根据不同的客户进行相应的促销活动
```
