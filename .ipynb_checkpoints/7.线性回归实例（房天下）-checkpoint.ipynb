```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('C:/Users/REGGIE/Desktop/数据分析资料/data/house_price_regression.xlsx')
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
#数据预处理
df['age'] = df['age'].map(lambda e:2020 -int(e.strip('建筑年代:')[1:5]))
df[['room','living_room']] = df['layout'].str.extract('(\d+)室(\d+)厅')
df['room'] = df['room'].astype('int')
df['living_room'] = df['living_room'].astype('int')
df['total_floor'] = df['floor_info'].str.extract('共(\d+)层')
df['total_floor'] = df['total_floor'].astype('int')
df['floor'] = df['floor_info'].str.extract('^(.)层')
df['direction'] = df['direction'].map(lambda e:e.strip())
```

```python
df.head()
```

```python
#清理没有用的数据
del df['url']
del df['title']
del df['floor_info']
del df['layout']
```

```python
df.head()
```

```python
#类别转换为dummy——variable
df = pd.concat([df,pd.get_dummies(df['direction']),pd.get_dummies(df['floor'])],axis = 1)
df.head()
```

```python
del df['南北向']
del df['低']
del df['direction']
del df['floor']
```

```python
df.head()
```

```python
#绘制散布图查看数据散布
%pylab inline
df[['price','area']].plot(kind='scatter',x='area',y='price',figsize=[10,5])
```

```python
#分析价格和每平方米的关系
#X用dataframe的方式取出
#目的：scikit-learn要求X是一个特征矩阵，y是一个NumPy向量
y = df['price']
x = df[['area']]
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x,y)
```

```python
print('Coefficient:{}'.format(regr.coef_))
print('Intercept:{}'.format(regr.intercept_))
```

```python
plt.scatter(x,y,color= 'blue')
plt.plot(x,regr.predict(x),linewidth=3,color='red')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
#多元回归预测
y = df['price'].values
x = df[['age', 'area', 'room', 'living_room', 'total_floor', '东南向','东向', '南向', '西南向', '西向', '中', '高']]
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x,y)
```

```python
x.info()
```

```python
#评估回归模型
import statsmodels.api as sm
x2 = sm.add_constant(x)
est = sm.OLS(y,x2)
est2 = est.fit()
print(est2.summary())
```

```python
df.columns
predictorcols = ['age', 'area', 'room', 'living_room', 'total_floor', '东南向','东向', '南向', '西南向', '西向', '中', '高']
```

```python
#透过AIC选定最佳的组合
import itertools
AICs = {}
for k in range(1,len(predictorcols)+1):
    for variables in itertools.combinations(predictorcols, k):
        predictors  = x[list(variables)]
        predictors2 = sm.add_constant(predictors.values)
        est = sm.OLS(y, predictors2)
        res = est.fit()
        AICs[variables] = res.aic
```

```python
from collections import Counter
c = Counter(AICs)
#c.most_common()
#去AIC最小的模型
c.most_common()[::-5]
```
