```python
def getinfodetail(url):
    info = {}
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    info['标题'] = soup.select('.house-title h1')[0].text
    info['价格'] = soup.select('.house-basic-item1 span')[0].text
    
    for i in soup.select('#generalSituation .general-item-left li'):
        #print(i.text.strip().split())
        key,value = i.text.strip().split()
        info[key] = value
    
    for i in soup.select('#generalSituation .general-item-right'):
       #print(i.text.strip().split())
        a = i.text.strip().split()
        del a[4]
        
    for i in range(0,len(a),2):
        b = a[i:i+2]
        info[b[0]] = b[1]
        
    return info
```

```python
#代码的整合
def getinfodetail(url):
    info = {}
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    info['标题'] = soup.select('.house-title h1')[0].text
    info['价格'] = soup.select('.house-basic-item1 span')[0].text
    
    for i in soup.select('#generalSituation .general-item-left li'):
        #print(i.text.strip().split())
        key,value = i.text.strip().split()
        info[key] = value
    
    for i in soup.select('#generalSituation .general-item-right'):
       #print(i.text.strip().split())
        a = i.text.strip().split()
        del a[4]
        
    for i in range(0,len(a),2):
        b = a[i:i+2]
        info[b[0]] = b[1]
        
    return info

import requests
from bs4 import BeautifulSoup
res = requests.get('https://sh.58.com/ershoufang/?PGTID=0d100000-0000-2f61-c63c-e0cefb57f787&ClickID=1')
soup = BeautifulSoup(res.text,'html.parser')
infoarray=[]
for item in soup.select('.house-list-wrap li'):
    title = item.select('.title a')[0].text.strip()
    url = item.select('.title a')[0]['href']
    print(url)
    infoarray.append(getinfodetail(url))
```
