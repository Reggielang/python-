# pythonIO与文件处理

```python
f = open('tmp.txt','w')
f.write('hello world')
f.close()
```

```python
with open('tmp.txt','w') as f:
    f.write('hello\nworld')
```

```python
with open('tmp.txt','r') as f:
    print(f.read())
```

```python
with open('tmp.txt','r') as f:
    for line in f.readlines():
        print(line.strip())
```
