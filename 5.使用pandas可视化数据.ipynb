```python
import pandas_datareader
```

```python
df = pandas_datareader.DataReader('BABA',data_source='yahoo',start = '2014-10-01')
df.head()
```

```python
%pylab inline
```

```python
#结合pandas 以及matplotlib 画图更加容易
df['Close'].plot(kind = 'line',figsize = [10,5], legend = True, title = 'BABA',grid = True)
```

```python
#df['mvg30']
df['mvg30'] = df['Close'].rolling(window = 30).mean()
```

```python
df.head()
```

```python
#折线图
df[['Close','mvg30']].plot(kind = 'line',figsize = [10,5])
```

```python
df.head()
```

```python
#直方图
df.loc[df.index > '2019-12-01','Volume'].plot(kind= 'bar', legend= True,title ='BABA')
```

```python
#圆饼图
#先取得开盘到收盘时的价差，然后取得每天是涨还是跌
df['diff'] = df['Close']-df['Open']
df['rise'] = df['diff'] > 0
df['fall'] = df['diff'] < 0
```

```python
df[['rise','fall']].sum().plot(kind = 'pie',figsize = [5,5], counterclock = False, startangle = 90, legend =True ,title = 'rise VS fall')
```
