```python
import pandas as pd
df = pd.read_csv('C:/Users/REGGIE/Desktop/数据分析资料/data/customer_churn.csv',header = 0,index_col = 0)
df.head()
```

```python
#数据预处理
df.columns
```

```python
#逗号前行，逗号后列,前三列没有实际用处，先去掉
df = df.iloc[:,3:]
```

```python
#如果直接使用dummy——variable对于只有yes和no的列来说，比较麻烦，所以直接使用map函数
cat_var = ['international_plan','voice_mail_plan','churn']

for v in cat_var:
    df[v] = df[v].map(lambda x : 1 if x == 'yes' else 0 )
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#使用决策树建立分类模型
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=5)
clf.fit(x,y)
```

```python
tree.export_graphviz(clf,out_file='tree_example.dot')
#打开命令行输入dot -Tpng tree_example.dot -o tree_example.png
```

```python
#模型的显示
%pylab inline
from IPython.display import Image
Image('tree_example.png')
```

```python
#第一个分类条件是x[3]，[15], [2]，所以最能预测客户是否流失的是这些值
x.columns[3]
```

```python
#检验结果
clf.predict(x)
```

```python
#查看准确度
import numpy as np
np.sum(y == clf.predict(x))/len(y)
```

```python
#使用逻辑回归
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression()
clf2.fit(x,y)
```

```python
clf2.predict(x)
np.sum(y == clf2.predict(x))/len(y)
#逻辑回归的准确率
```

```python
#使用SVM
from sklearn.svm import SVC
model = SVC()
model.fit(x,y)
```

```python
model.predict(x)
np.sum(y == model.predict(x))/len(y)
#SVM的准确度
```
