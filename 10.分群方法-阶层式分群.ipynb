```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendrogram = sch.dendrogram(sch.linkage(iris.data,method= 'ward'))
plt.title('Dendrogram')
plt.xlabel('Iris')
plt.ylabel('Eucliden distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(iris.data)

plt.scatter(iris.data[y_hc ==0,2],iris.data[y_hc == 0,3], s = 100, c ='red',label = 'C1')
plt.scatter(iris.data[y_hc ==1,2],iris.data[y_hc == 1,3], s = 100, c ='blue',label = 'C2')
plt.scatter(iris.data[y_hc ==2,2],iris.data[y_hc == 2,3], s = 100, c ='green',label = 'C3')
plt.title('Clusters of iris')
plt.xlabel('petal.length')
plt.ylabel('petal.width')
plt.legend()
plt.show()
```
