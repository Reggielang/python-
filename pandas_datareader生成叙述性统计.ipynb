```python
import pandas_datareader
```

```python
df = pandas_datareader.DataReader('BABA',data_source='yahoo')
```

```python
df.tail()
```

```python
df.columns
```

```python
df['Close'].sum()
df['Close'].mean()
df['Close'].std()
df['Close'].min()
df['Close'].max()
df['Close'].count()
```

```python
df[['Open','Close']].min()
df[['Open','Close']].max()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['diff'] = df['Close']-df['Open']
```

```python
df['rise'] = df['diff'] > 0 
```

```python
df['fall'] = df['diff'] < 0
```

```python
df.head()
```

```python
df[['rise','fall']].sum()
```

```python
df[df.index > '2017-04-01']
```

```python
#df.loc[ , ]逗号前放列的内容，逗号后放行的内容
df.loc[df.index > '2017-04-01' , ['rise','fall']].sum()
```

```python
df.index.year
df.index.month
```

```python
#根据年和月提取数据
df.groupby([df.index.year,df.index.month])['rise','fall'].sum()#.unstack() #变为长表格
```

```python
df['ret'] = df['Close'].pct_change(1)
```

```python
df.head()
```

```python
df.groupby([df.index.year,df.index.month])['ret'].mean().unstack()
```
