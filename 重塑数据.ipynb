```python
#虚拟变量的建立
import pandas as pd
df = pd.read_excel('C:/Users/REGGIE/Desktop/数据分析资料/data/house_sample.xlsx')
df.head()
```

```python
#建立了南北为1，南为0的两列虚拟变量
pd.get_dummies(df['朝向'])
```

```python
#合并这两列的虚拟变量
df = pd.concat([df, pd.get_dummies(df['朝向'])] ,axis = 1)
```

```python
df.head()
```

```python
df = df.drop('朝向',axis = 1)
df.head()
```

```python
#建立透视表
df2 = df.pivot_table(index = '张贴日期', columns = '产权性质', values = '总价', aggfunc = sum, fill_value = 0)
df2.head()
```

```python
df3 = df.pivot_table(index = '产权性质', columns = '张贴日期', values = '总价', aggfunc = sum, fill_value = 0)
df3.head()
```

```python
#转换index和columns
df2.T
```

```python
#建立多索引的透视表
df_multi_idx = df.pivot_table(index = ['装修','楼层'], columns = '产权性质', values = '总价', aggfunc = sum, fill_value = 0)
df_multi_idx
```

```python
#长表格转化为宽表格
df_wide = df_multi_idx.unstack()
df_wide.head()
```

```python
#宽表格转化为长表格
df_long = df_wide.stack()
df_long.head()
```
